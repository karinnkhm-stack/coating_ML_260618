## smiles_260616.csvからsmiles_list_merged.csvを作成

### Step1

```python
import pandas as pd

df_step1 = pd.read_csv(r"C:\Users\scarl\Downloads\大学関連\25講義関連\b3後期\b4\coating_ML2\coating_ML_260618\data\raw\smiles_260616.csv")

df_step1
```

### Step2

```python
import pandas as pd

parts = []
for i in range(1, 26):
    tmp = df_step1[["id", "name", f"smiles_{i}", f"ratio_{i}"]].copy()
    tmp = tmp.rename(columns={"id": "material_id", f"smiles_{i}": "smiles", f"ratio_{i}": "ratio"})
    tmp["component_no"] = i  # 成分順を保持したい場合（不要なら消してOK）
    parts.append(tmp)

df_step2 = pd.concat(parts, ignore_index=True)

df_step2["smiles"] = df_step2["smiles"].replace(r"^\s*$", pd.NA, regex=True)
df_step2["ratio"] = pd.to_numeric(df_step2["ratio"], errors="coerce")
df_step2 = df_step2.dropna(subset=["smiles", "ratio"]).reset_index(drop=True)

# 水を追加
df_step2 = pd.concat(
    [df_step2, pd.DataFrame([{"material_id": 144, "name": "water", "smiles": "O", "ratio": 100.0, "component_no": 0}])],
    ignore_index=True,
)

# 先に material_id 順（＋成分順）に並べる
df_step2 = df_step2.sort_values(["material_id", "component_no"], kind="stable").reset_index(drop=True)

# 並べた後に id を振る
df_step2.insert(0, "id", range(1, len(df_step2) + 1))

df_step2 = df_step2.drop(columns=["component_no"])
df_step2.to_csv("../../data/interim/smiles_list_merged.csv", index=False)

df_step2
```
