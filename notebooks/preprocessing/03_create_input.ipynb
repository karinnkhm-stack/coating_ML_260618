## structures.csv作成

```python
from pathlib import Path
import pandas as pd

in_path = Path("../../data/interim/smiles_list_merged.csv")
out_path = Path("../../data/interim/structures.csv")

df = pd.read_csv(in_path)

smiles = (
    df["smiles"]
    .astype(str)
    .str.strip()
)

structures = (
    smiles[smiles.ne("") & smiles.ne("nan")]
    .drop_duplicates(keep="first")
    .reset_index(drop=True)
    .to_frame(name="smiles")
)

structures.insert(0, "id", range(1, len(structures) + 1))

structures.to_csv(out_path, index=False, encoding="utf-8-sig")
```
