## strutures_basic_descriptors.csvを作成
### RDkit記述子生成

```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

in_path = "../../data/interim/structures.csv"
out_path = "../../data/interim/structures_basic_descriptors.csv"

df = pd.read_csv(in_path)

# 列名ゆれ対応（念のため）
if "smiles" not in df.columns and "smile" in df.columns:
    df = df.rename(columns={"smile": "smiles"})

descriptor_names = [
    "ExactMolWt", "MolLogP", "NumHDonors", "NumHAcceptors", "TPSA",
    "NumRotatableBonds", "FractionCSP3", "LabuteASA", "BalabanJ",
    "HeavyAtomCount", "NOCount", "NumAromaticRings"
]

desc_map = dict(Descriptors._descList)
missing = [name for name in descriptor_names if name not in desc_map]
if missing:
    raise ValueError(f"RDKitに存在しない記述子名があります: {missing}")

selected_desc_list = [(name, desc_map[name]) for name in descriptor_names]

def smiles_to_mol(s):
    if pd.isna(s):
        return None
    return Chem.MolFromSmiles(str(s))

def calc_desc(m):
    if m is None:
        return {name: np.nan for name in descriptor_names}
    return {name: fn(m) for name, fn in selected_desc_list}

mols = df["smiles"].map(smiles_to_mol)
desc_df = pd.DataFrame([calc_desc(m) for m in mols], index=df.index)

out_df = pd.concat([df[["id", "smiles"]], desc_df], axis=1)
out_df.to_csv(out_path, index=False)
```
