## 01．データ読み込み・WVPの外れ値除外（0～4のみに）

```python
import pandas as pd

def wvp_clean_data(csv_path):
    print("--- Step 1: データの読み込みと外れ値の除外 ---")
    df = pd.read_csv(csv_path)

    # WVP：0〜4の範囲のみを抽出
    wvp_filter = (df['wvp'] >= 0) & (df['wvp'] <= 4)
    df_cleaned = df[wvp_filter].copy()

    print(f"生のサンプル数: {len(df)} 件 -> 除外後のサンプル数: {len(df_cleaned)} 件")

    # ステップ1の成果を保存
    df_cleaned.to_csv(r"C:\Users\scarl\Downloads\大学関連\25講義関連\b3後期\b4\coating_ML2\coating_ML_260618\data\interim\260625\merged_basic_descriptors.csv", index=False)
    print("Step 1 のデータを保存しました。")
    return df_cleaned

#実行用
if __name__ == "__main__":
    path = r"C:\Users\scarl\Downloads\大学関連\25講義関連\b3後期\b4\coating_ML2\coating_ML_260618\data\interim\260625\merged_basic_descriptors.csv"
    df_clean = wvp_clean_data(path)
```
