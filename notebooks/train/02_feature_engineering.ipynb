## 02．SMILESを数値に翻訳
可食コーティングの「水蒸気の通しやすさ」に深く関わる21個の記述子（親水性、水素結合数、立体障害など）をRDKitで計算

※読み込めないSMILESはNaN（欠損値）扱い

```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

df_clean = pd.read_csv(r"../data/interim/01_cleaned_raw.csv")

def calculate_selected_descriptors(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return [np.nan] * 21
    return [
        Descriptors.ExactMolWt(mol), Descriptors.MolLogP(mol), Descriptors.NumHDonors(mol),
        Descriptors.NumHAcceptors(mol), Descriptors.TPSA(mol), Descriptors.NumRotatableBonds(mol),
        Descriptors.FractionCSP3(mol), Descriptors.LabuteASA(mol), Descriptors.BalabanJ(mol),
        Descriptors.HeavyAtomCount(mol), Descriptors.NumAromaticRings(mol), Descriptors.MolMR(mol),
        Descriptors.Kappa1(mol), Descriptors.Kappa2(mol), Descriptors.HallKierAlpha(mol),
        Descriptors.NumHeteroatoms(mol), Descriptors.HeavyAtomMolWt(mol), Descriptors.NHOHCount(mol),
        Descriptors.NOCount(mol), Descriptors.RingCount(mol), Descriptors.MaxAbsPartialCharge(mol)
    ]

def get_descriptor_dataframe(df):
    print("--- Step 2: 分子記述子の計算 ---")
    descriptor_names = [
        "ExactMolWt", "MolLogP", "NumHDonors", "NumHAcceptors", "TPSA",
        "NumRotatableBonds", "FractionCSP3", "LabuteASA", "BalabanJ", "HeavyAtomCount",
        "NumAromaticRings", "MolMR", "Kappa1", "Kappa2", "HallKierAlpha",
        "NumHeteroatoms", "HeavyAtomMolWt", "NHOHCount", "NOCount", "RingCount", "MaxAbsPartialCharge"
    ]
    print("SMILESから化学構造特徴を抽出中...")
    desc_values = [calculate_selected_descriptors(s) for s in df['smiles']]
    return pd.DataFrame(desc_values, columns=descriptor_names)


#実行用
if __name__ == "__main__":
    import pandas as pd
    import os

    import_path = r"C:\Users\scarl\Downloads\大学関連\25講義関連\b3後期\b4\coating_ML2\coating_ML_260618\data\interim\01_cleaned_raw.csv"
    df_clean = pd.read_csv(import_path)
    df_desc = get_descriptor_dataframe(df_clean)

    export_path = r"C:\Users\scarl\Downloads\大学関連\25講義関連\b3後期\b4\coating_ML2\coating_ML_260618\data\interim\02_molecular_descriptors.csv"
    df_desc.to_csv(export_path, index=False)

    print("Step 2 の分子記述子を保存しました！")
```
